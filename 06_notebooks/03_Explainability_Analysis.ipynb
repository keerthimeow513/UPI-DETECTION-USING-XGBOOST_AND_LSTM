# Explainability Analysis (XAI) using SHAP

Understanding why the model classifies certain transactions as fraud using SHAP (SHapley Additive exPlanations).

```python
import sys
sys.path.append('..')

import shap
import joblib
import matplotlib.pyplot as plt
import numpy as np
from utils.preprocessing import Preprocessor
```

```python
# Load Model and Data
xgb_model = joblib.load('../02_models/artifacts/xgb_model.pkl')
preprocessor = Preprocessor('../07_configs/config.yaml')
preprocessor.load_artifacts()

# Load a sample for explanation
# Note: In a real notebook we would load X_xgb_train from disk if saved, or regenerate.
# Here we regenerate briefly.
import pandas as pd
df = pd.read_csv('../01_data/raw/upi_transactions.csv').head(1000)
df_processed = preprocessor.fit_transform(df)
_, X_xgb, _ = preprocessor.create_sequences(df_processed)

print("Feature Names:", preprocessor.feature_names)
```

```python
# Initialize Explainer
explainer = shap.TreeExplainer(xgb_model)
shap_values = explainer.shap_values(X_xgb)

# Summary Plot
shap.summary_plot(shap_values, X_xgb, feature_names=preprocessor.feature_names)
```
